```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
from math import exp
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
sx = preprocessing.MinMaxScaler()
sy = preprocessing.MinMaxScaler()
```

```python
df.shape[0]
```

```python
scaled_x = sx.fit_transform(df.drop('bought_insurance', axis = 'columns'))
scaled_x
```

```python
scaled_y = sy.fit_transform(df['bought_insurance'].values.reshape(df.shape[0],1))
scaled_y.shape[0]
```

```python
scaled_y.reshape(28,)
```

```python
scaled_x[0][0]
```

```python
x_train ,x_test, y_train, y_test = train_test_split(scaled_x,scaled_y, test_size = 0.2, random_state = 42)
```

```python
size = x_train.shape[0]
```

```python
def sig_func(a):
    return 1/(1+np.exp(-a))
```

```python
def sgdd(x,y,epochs,lr):
    w1=w2=1
    bias = 0
    for i in range(epochs):
        idx = random.randint(0,size-1)
        ws = w1 * x[idx][0] + w2 * x[idx][1] + bias
        ws = np.squeeze(ws)
        y_pred = sig_func(ws)
        cost = (y[idx] - y_pred)**2
        w1d = -2 * (y[idx] - y_pred)*x[idx][0]
        w2d = -2 * (y[idx] - y_pred)* x[idx][1]
        bd = -2 * (y[idx] - y_pred)
        w1 = w1 - lr * w1d
        w2 = w2 - lr*w2d
        bias = bias - lr*bd
    return cost,w1,w2,bias
    
    
```

```python
cost,w1,w2,bias = sgdd(x_train,y_train,500,0.01)
cost,w1,w2,bias
```
